```python
df.sex = pd.Categorical(df.sex.str.strip())
df.relationship = pd.Categorical(df.relationship.str.strip())
df.race = pd.Categorical(df.race.str.strip())
df.label = pd.Categorical(df.label.str.strip())
df = df.loc[df['workclass'] != '?',:]
df = df.loc[df['occupation'] != '?',:]
df = df.loc[df['native country'] != '?',:]
df
```

```python
from sklearn.model_selection import train_test_split
X = pd.get_dummies(df.iloc[:,0:13])
y = df.label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=15)
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix

y_predict = knn.predict(X_test)
y_predict
```

```python
cnf = confusion_matrix(y_test, y_predict, labels=['No','Yes'])
cnf
```

```python
sns.heatmap(pd.DataFrame(cnf, index=['No','Yes'], 
                         columns=['No','Yes']), annot=True, cmap='YlGnBu')
```

```python
from sklearn.metrics import precision_recall_fscore_support

pre, rec, fs, sup = precision_recall_fscore_support(y_test, y_predict, 
                                                    labels=['No','Yes'])
pre
```
