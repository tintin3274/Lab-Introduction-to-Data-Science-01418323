## 20.

```python
df20 = answer19.copy()
df20.reset_index(inplace=True)
df20 = df20.fillna(0.0)
```

```python
b20 = df20.parent_incident_type.isin(['Property Crime', 'Breaking & Entering', 'Drugs'])
```

```python
df20_1 = df20.loc[b20].groupby(['zip', 'parent_incident_type']).size().to_frame(name='Ratio')
df20_1.reset_index(inplace=True)
df20_1
```

```python
df20_2 = df20.loc[b20].groupby(['parent_incident_type']).size().to_frame(name='Amount')
df20_2
```

```python
df20_1.join(df20_2, on='parent_incident_type')
```

```python
df20_1.Ratio = df20_1.Ratio / df20_1.join(df20_2, on='parent_incident_type').Amount
answer20 = df20_1.copy()
answer20
```

## 21.

```python
df21_1 = df20_1.copy()
df21_1.zip = df21_1.zip.astype(int)
df21_1
```

```python
df21_2 = df21_1.pivot_table(values='Ratio', index='zip', columns='parent_incident_type', aggfunc='first')
df21_2 = df21_2.fillna(0.0)
df21_2
```

```python
house_sales = house_sales.join(df21_2, on='zipcode')
```

```python
house_sales.loc[:,'Breaking & Entering'] = house_sales.loc[:,'Breaking & Entering'].fillna(0.0)
house_sales.loc[:,'Drugs'] = house_sales.loc[:,'Drugs'].fillna(0.0)
house_sales.loc[:,'Property Crime'] = house_sales.loc[:,'Property Crime'].fillna(0.0)
house_sales
```
