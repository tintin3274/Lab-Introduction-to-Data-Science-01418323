```python
b24 = (medals.W_Gold + medals.W_Silver + medals.W_Bronze) > (medals.S_Gold + medals.S_Silver + medals.S_Bronze)
```

```python
Total_Medals = pd.DataFrame({
    'Total_Medals_Summer': medals.S_Gold + medals.S_Silver + medals.S_Bronze, 
    'Total_Medals_Winter': medals.W_Gold + medals.W_Silver + medals.W_Bronze})
```

```python
Total_Medals
```

```python
medals.loc[b24, ['Country']]
```

```python
Total_Medals.loc[b24, :]
```

```python
pd.concat([medals.loc[b24, ['Country']], Total_Medals.loc[b24, :]], axis=1)
```

## 25.

```python
b25 = ((Total_Medals.Total_Medals_Summer == 0) & (Total_Medals.Total_Medals_Winter > 0)) | ((Total_Medals.Total_Medals_Summer > 0) & (Total_Medals.Total_Medals_Winter == 0))
```

```python
pd.concat([medals.loc[b25, ['Country']], Total_Medals.loc[b25, :]], axis=1)
```

## 29.

```python
a29 = (((medals.S_Gold*3) + (medals.S_Silver*2) + medals.S_Bronze) / medals.No_Summer).fillna(0)
```

```python
b29 = (((medals.W_Gold*3) + (medals.W_Silver*2) + medals.W_Bronze) / medals.No_Winter).fillna(0)
```

```python
medals['perGame'] = a29 + b29
```

```python
medals.sort_values('perGame')[0:10]
```
