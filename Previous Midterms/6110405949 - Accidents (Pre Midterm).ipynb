## 25.

```python
prov_area = pd.read_csv('provSize.csv')
prov_area.columns = ['Rank', 'Provinde', 'Area']
prov_area = prov_area.drop('Rank', axis=1)
prov_area.Area = prov_area.Area.str.replace(',', '').str.extract('(\d*\.\d+|\d+)', expand=False).astype(float)
prov_area.iloc[68,0] = 'กรุงเทพมหานคร'
prov_area = prov_area.set_index('Provinde')
prov_area
```

```python
accidents = accidents.join(prov_area, on='Provinde')
accidents = accidents.set_index(['Provinde', 'Event'])
accidents = accidents.sort_index(level=['Provinde', 'Event'], ascending=[True, True])
accidents
```

## 26.

```python
df26_1 = accidents.groupby(['Provinde', 'Event']).size().to_frame(name='Counts').reset_index()
df26_1
```

```python
df26_2 = df26_1.join(prov_area, on='Provinde')
df26_2['Dead_Per_Area'] = df26_2['Counts']/df26_2['Area']
df26_2
```

```python
df26_3 = df26_2.loc[df26_2.Event == 'ปีใหม่ 58']
df26_3 = df26_3.sort_values('Dead_Per_Area', ascending=False)
answer26 = df26_3.head(5)
answer26
```
